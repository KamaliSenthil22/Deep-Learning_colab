```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import tensorflow as tf
from tensorflow import keras
```

```python
iris_df=load_iris()
```

```python
iris_df.keys()
```

```python
x=pd.DataFrame(data=iris_df["data"],columns=iris_df["feature_names"])
x
```

```python
y=pd.DataFrame(data=iris_df['target'])
y
```

#Standardization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_model=MinMaxScaler()
```

```python
scaled_x=scaler_model.fit_transform(x)
```

```python
scaled_x=pd.DataFrame(scaled_x,columns=x.columns)
```

# Model Validation technique

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.20, random_state=42)
```

```python
X_train
```

# Model Building

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
model=Sequential()
model.add(Dense(units=20,input_dim=4,activation="relu",kernel_initializer="glorot_uniform"))
model.add(Dense(units=15,activation="relu",kernel_initializer="glorot_uniform"))
model.add(Dense(units=3,activation="softmax",kernel_initializer="glorot_uniform"))
model.summary()
```

#Model Compile

```python
adam_opt=tf.keras.optimizers.Adam(
    learning_rate=0.001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07)
```

```python
model.compile(optimizer=adam_opt,loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

# Model Training

```python
model_train=model.fit(X_train,y_train,epochs=100,batch_size=20,verbose=1)
```

```python
def create_model(neuron1,neuron2,activation_function,init,learning_rate,Dropout_rate):
  model=Sequential()
  model.add(Dense(units=neuron1,input_dim=4,activation=activation_function,kernel_initializer=init))
  model.add(Dropout(Dropout_rate))
  model.add(Dense(units=neuron2,activation=activation_function,kernel_initializer=init))
  model.add(Dropout(Dropout_rate))
  model.add(Dense(units=3,activation="softmax",kernel_initializer="glorot_uniform"))
  model.summary()

  adam_opt=tf.keras.optimizers.Adam(
    learning_rate=learning_rate,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07)
  model.compile(optimizer=adam_opt,loss="sparse_categorical_crossentropy",metrics=["accuracy"])

  return model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout,Input
from tensorflow.keras.optimizers import Adam
from scikeras.wrappers import KerasClassifier


def create_model(neuron1=10, neuron2=10, activation_function="relu",
                 init="uniform", learning_rate=0.01, Dropout_rate=0.1):
    model = Sequential()
    model.add(Input(shape=(8,)))
    model.add(Dense(units=neuron1, activation=activation_function, kernel_initializer=init))
    model.add(Dropout(Dropout_rate))
    model.add(Dense(units=neuron2, activation=activation_function, kernel_initializer=init))
    model.add(Dropout(Dropout_rate))
    model.add(Dense(units=1, activation='sigmoid', kernel_initializer='glorot_uniform'))

    adam_opt = Adam(learning_rate=learning_rate, beta_1=0.5, beta_2=0.96, epsilon=1e-07)
    model.compile(optimizer=adam_opt, loss="binary_crossentropy", metrics=["accuracy"])
    return model

# Parameter grid

param_grid = {
    'model__neuron1': [10, 20, 10],
    'model__neuron2': [10, 20, 10],
    'model__activation_function': ["relu", "tanh"],
    'model__init': ["uniform", "zeros"],
    'model__learning_rate': [0.1, 0.01, 0.001],
    'model__Dropout_rate': [0.1, 0.3],
    'batch_size': [10, 20, 30],
    'epochs':[5,10,12]}



model = KerasClassifier(model=create_model, verbose=0)
grid_result = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result.fit(X_train,y_train)
```
